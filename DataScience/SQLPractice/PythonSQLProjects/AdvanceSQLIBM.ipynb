### Join Section

```python
# Write and execute a SQL query to list the school names, community names and average attendance 
# for communities with a hardship index of 98.

%sql SELECT SD.NAME_OF_SCHOOL as 'SchoolNames', CD.COMMUNITY_AREA_NAME as 'CommAreaName', \
SD.AVERAGE_STUDENT_ATTENDANCE as 'AvgAttendance' FROM SchoolData as SD \
LEFT JOIN CensusData as CD ON SD.COMMUNITY_AREA_NUMBER = CD.COMMUNITY_AREA_NUMBER \
WHERE HARDSHIP_INDEX = 98;
```

```python
# Write and execute a SQL query to list all crimes that took place at a school. 
# Include case number, crime type and community name.

%sql SELECT CD.CASE_NUMBER as 'CaseNumber', CD.PRIMARY_TYPE as 'CrimeType', CSD.COMMUNITY_AREA_NAME as 'CommAreaName' \
FROM CrimeData as CD LEFT JOIN CensusData as CSD ON CD.COMMUNITY_AREA_NUMBER = CSD.COMMUNITY_AREA_NUMBER\
WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%' \
ORDER BY LENGTH(CrimeType) DESC;
```

### Creating a view

For privacy reasons, you have been asked to create a view that enables users to select just the school name and the icon fields from the CHICAGO_PUBLIC_SCHOOLS table. By providing a view, you can ensure that users cannot see the actual scores given to a school, just the icon associated with their score. You should define new names for the view columns to obscure the use of scores and icons in the original table.

```python
# Write and execute a SQL statement to create a view.

%sql CREATE VIEW ChicagoPublicSchoolsView AS\
    SELECT SD.NAME_OF_SCHOOL as 'School_name', SD.Safety_Icon as 'Safety_Rating',\
    SD.Family_Involvement_Icon as 'Family_Rating', SD.Environment_Icon as 'Environment_Rating',\
    SD.Instruction_Icon as 'Instruction_Rating', SD.Leaders_Icon as 'Leaders_Rating',\
    SD.Teachers_Icon as 'Teachers_Rating'\
    FROM SchoolData as SD;
```

```python

# Define function to be able to run precedures in sqlite3 and jupyternotebooks
def update_leaders_procedure(school_ID, leaders_score):
    # Update Leaders_Score
    cur.execute(f"UPDATE SchoolData SET Leaders_Score = {leaders_score} WHERE School_ID = {school_ID}")
    
    # Actualiza Leaders_Icon
    cur.execute(f'''
        UPDATE SchoolData
        SET Leaders_Icon = CASE
            WHEN Leaders_Score > 0 AND Leaders_Score < 19 THEN 'Very Weak'
            WHEN Leaders_Score < 40 THEN 'Weak'
            WHEN Leaders_Score < 60 THEN 'Medium'
            WHEN Leaders_Score < 80 THEN 'Strong'
            WHEN Leaders_Score < 100 THEN 'Strong'
             
        END
        WHERE School_ID = {school_ID}
    ''')

# Luego, puedes llamar a la funcin con los argumentos deseados
update_leaders_procedure(610038, 82)
```
